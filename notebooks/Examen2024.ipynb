### Coneixements previs

- Per convertir una temperatura $T_C$ en graus centígrads a una temperatura $T_K$ en graus Kelvin:
    - $T_C = T_K + 273.15$

- Primer principi de la termodinàmica:
    - $\frac{d U}{d t} = Q - W$
    - U és l'energia interna del sistema
    - Q és el flux calorífic total en el sistema
    - W potència desenvolupada pel sistema

- Calor específic $c_e$:
    - $c_e = \frac{1}{m} \frac{d U}{d T}$
    - m és la massa del sistema
    - U és l'energia interna del sistema
    - T és la temperatura del sistema
    
- Conducció tèmica
    - $Q_x = - k A \frac{d T}{d x}$
    - $Q_x$ és el flux de calor en la direcció $x$
    - k és la conductivitat del material
    - A és l'àrea de la secció a través de la qual flueix la calor
    
- Radiació tèrmica
    - Tot cos calent radia energia segons la llei d'Stefan–Boltzmann
    - $Q_r = \sigma \epsilon A T^4$
    - $Q_r$ és el flux de calor per radiació d'un cos a temperatura T
    - $\sigma = 5.67040 \times 10^{−8} \pm 4 \times 10^{−13}$ $W/(m^2 K^4)$ és la constant d'Stefan–Boltzmann
    - A és la superfície de l'àrea radiant
    - T és la temperatura del cos radiant
    
- Radiació solar
    - El sol radia amb una potència de $1420 \pm 50$ $W/m^2$

- Conductivitat de diferents materials [W/(m K)]
    - Aire: $0.026 \pm 0.005$ 
    - Fibra de carboni: $4.0 \pm 0.5$ 
    - Coure: $384.1 \pm 0.5$
    - Alumini: $183 \pm 1$
